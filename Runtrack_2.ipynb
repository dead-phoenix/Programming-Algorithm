```python
class Person:
    def __init__(self, name, firstname):
        self.__name = name  # Using double underscores to make these attributes private
        self.__firstname = firstname

    def SePresenter(self):
        print(f"My name is {self.__firstname} {self.__name}.")

    # Accessor (getter) methods
    def get_name(self):
        return self.__name

    def get_firstname(self):
        return self.__firstname

    # Mutator (setter) methods
    def set_name(self, name):
        self.__name = name

    def set_firstname(self, firstname):
        self.__firstname = firstname

# Instantiate some Person objects
person1 = Person("Doe", "John")
person2 = Person("Smith", "Alice")

# Use the SePresenter method to verify
person1.SePresenter()
person2.SePresenter()

# Access and modify attributes using accessor and mutator methods
print("Original name:", person1.get_name())
print("Original firstname:", person1.get_firstname())

person1.set_name("Brown")
person1.set_firstname("Michael")

print("Modified name:", person1.get_name())
print("Modified firstname:", person1.get_firstname())
```

```python
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author

    def print(self):
        print(self.title)


class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name


class Author(Person):
    def __init__(self, first_name, last_name):
        super().__init__(first_name, last_name)
        self.work = []

    def listerWork(self):
        print(f"Books written by {self.first_name} {self.last_name}:")
        for book in self.work:
            print(book.title)

    def writeABook(self, title):
        new_book = Book(title, self)
        self.work.append(new_book)
        return new_book


# Example usage:
if __name__ == "__main__":
    author1 = Author("John", "Doe")
    author2 = Author("Jane", "Smith")

    book1 = author1.writeABook("Book 1 by John Doe")
    book2 = author2.writeABook("Book 1 by Jane Smith")
    book3 = author1.writeABook("Book 2 by John Doe")

    author1.listerWork()
    author2.listerWork()
```

```python
class Person:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name


class Author(Person):
    def __init__(self, first_name, last_name):
        super().__init__(first_name, last_name)
        self.work = []

    def listerWork(self):
        return [book.title for book in self.work]

    def writeABook(self, title):
        new_book = Book(title, self)
        self.work.append(new_book)
        return new_book


class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author


class Library:
    def __init__(self, name):
        self.name = name
        self.catalog = {}

    def buyBook(self, author, title, quantity):
        for book in author.work:
            if book.title == title:
                if title in self.catalog:
                    self.catalog[title] += quantity
                else:
                    self.catalog[title] = quantity

    def inventory(self):
        print("Library Catalog:")
        for title, quantity in self.catalog.items():
            print(f"{title} - Quantity: {quantity}")

    def rent(self, customer, title):
        if title in self.catalog and self.catalog[title] > 0:
            if title not in customer.books:
                customer.books.append(title)
            self.catalog[title] -= 1
            return True
        return False

    def renderBooks(self, customer):
        for title in customer.books:
            if title in self.catalog:
                self.catalog[title] += 1


class Customer(Person):
    def __init__(self, first_name, last_name):
        super().__init__(first_name, last_name)
        self.books = []

    def inventory(self):
        print(f"{self.first_name} {self.last_name}'s Book Collection:")
        for title in self.books:
            print(title)


# Instantiate authors and have them write books
author1 = Author("John", "Doe")
book1 = author1.writeABook("Book 1 by John Doe")
book2 = author1.writeABook("Book 2 by John Doe")

author2 = Author("Jane", "Smith")
book3 = author2.writeABook("Book 1 by Jane Smith")

# Create a library, have it buy books
library = Library("City Library")
library.buyBook(author1, "Book 1 by John Doe", 5)
library.buyBook(author1, "Book 2 by John Doe", 3)
library.buyBook(author2, "Book 1 by Jane Smith", 2)

# Create customers, have them rent books
customer1 = Customer("Alice", "Johnson")
customer2 = Customer("Bob", "Smith")

library.rent(customer1, "Book 1 by John Doe")
library.rent(customer1, "Book 1 by Jane Smith")
library.rent(customer2, "Book 2 by John Doe")

# Display inventory of the library and customers
library.inventory()
customer1.inventory()
customer2.inventory()

# Render books back to the library
library.renderBooks(customer1)
library.renderBooks(customer2)

# Display updated inventory
library.inventory()
```

```python
class Board:
    def __init__(self, i, j):
        self.rows = i
        self.columns = j
        self.board = [['O' for _ in range(j)] for _ in range(i)]

    def print_board(self):
        for row in self.board:
            print(' '.join(row))
        print('-' * (self.columns * 2 - 1))
        for col in range(self.columns):
            print(f'{col + 1}', end=' ')
        print()

    def play(self, column, color):
        if color != 'R' and color != 'J':
            print("Invalid color. Use 'R' for red or 'J' for yellow.")
            return False

        if column < 1 or column > self.columns:
            print("Invalid column. Please choose a valid column.")
            return False

        for i in range(self.rows - 1, -1, -1):
            if self.board[i][column - 1] == 'O':
                self.board[i][column - 1] = color
                return True

        print("Column is full. Please choose another column.")
        return False

    def check_winner(self, color):
        # Check horizontal
        for row in self.board:
            if f'{color * 4}' in ''.join(row):
                return True

        # Check vertical
        for col in range(self.columns):
            column = ''.join([self.board[row][col] for row in range(self.rows)])
            if f'{color * 4}' in column:
                return True

        # Check diagonals (from top-left to bottom-right)
        for i in range(self.rows - 3):
            for j in range(self.columns - 3):
                diagonal = ''.join([self.board[i + k][j + k] for k in range(4)])
                if f'{color * 4}' in diagonal:
                    return True

        # Check diagonals (from top-right to bottom-left)
        for i in range(3, self.rows):
            for j in range(self.columns - 3):
                diagonal = ''.join([self.board[i - k][j + k] for k in range(4)])
                if f'{color * 4}' in diagonal:
                    return True

        return False


def main():
    i = int(input("Enter the number of rows: "))
    j = int(input("Enter the number of columns: "))
    board = Board(i, j)
    player_turn = 'R'  # Red starts

    while True:
        board.print_board()
        column = int(input(f"Player {player_turn}, choose a column (1-{j}): "))
        if board.play(column, player_turn):
            if board.check_winner(player_turn):
                board.print_board()
                print(f"Player {player_turn} wins! You receive 100,000 euros.")
                break
            player_turn = 'J' if player_turn == 'R' else 'R'


if __name__ == "__main__":
    main()
```

```python
import random

class AI_One:
    def __init__(self, color):
        self.color = color

    def think(self, board):
        # This AI randomly selects an available column to play
        available_columns = [col for col in range(board.columns) if board.board[0][col] == 'O']
        return random.choice(available_columns) + 1  # Add 1 to convert to 1-based indexing

class Player:
    def __init__(self, name, color):
        self.name = name
        self.color = color

class ConnectFourGame:
    def __init__(self, rows, columns):
        self.board = [['O' for _ in range(columns)] for _ in range(rows)]
        self.rows = rows
        self.columns = columns
        self.current_player = None

    def print_board(self):
        for row in self.board:
            print(' '.join(row))
        print('-' * (self.columns * 2 - 1))
        for col in range(self.columns):
            print(f'{col + 1}', end=' ')
        print()

    def is_valid_move(self, column):
        return 1 <= column <= self.columns and self.board[0][column - 1] == 'O'

    def play(self, player, column):
        for i in range(self.rows - 1, -1, -1):
            if self.board[i][column - 1] == 'O':
                self.board[i][column - 1] = player.color
                return True
        return False

    def check_winner(self, player):
        color = player.color
        for i in range(self.rows):
            for j in range(self.columns):
                if (
                    j + 3 < self.columns and self.board[i][j] == self.board[i][j + 1] == self.board[i][j + 2] == self.board[i][j + 3] == color
                ) or (
                    i + 3 < self.rows and self.board[i][j] == self.board[i + 1][j] == self.board[i + 2][j] == self.board[i + 3][j] == color
                ) or (
                    i + 3 < self.rows and j + 3 < self.columns and self.board[i][j] == self.board[i + 1][j + 1] == self.board[i + 2][j + 2] == self.board[i + 3][j + 3] == color
                ) or (
                    i - 3 >= 0 and j + 3 < self.columns and self.board[i][j] == self.board[i - 1][j + 1] == self.board[i - 2][j + 2] == self.board[i - 3][j + 3] == color
                ):
                    return True
        return False

    def is_board_full(self):
        return all(cell != 'O' for row in self.board for cell in row)

    def play_game(self):
        player1 = Player("Human", "R")
        ai = AI_One("Y")  # AI plays with yellow tokens

        while True:
            self.print_board()

            if self.current_player == player1:
                while True:
                    try:
                        column = int(input(f"{self.current_player.name}, choose a column (1-{self.columns}): "))
                        if self.is_valid_move(column):
                            break
                        else:
                            print("Invalid move. Please choose a valid column.")
                    except ValueError:
                        print("Invalid input. Please enter a valid column number.")
            else:
                print("AI is thinking...")
                column = ai.think(self)

            if self.play(self.current_player, column):
                if self.check_winner(self.current_player):
                    self.print_board()
                    print(f"{self.current_player.name} ({self.current_player.color}) wins!")
                    break
                elif self.is_board_full():
                    self.print_board()
                    print("It's a tie!")
                    break

            self.current_player = player1 if self.current_player != player1 else ai

if __name__ == "__main__":
    rows = 6
    columns = 7
    game = ConnectFourGame(rows, columns)
    game.current_player = random.choice([Player("Human", "R"), AI_One("Y")])
    game.play_game()
```
